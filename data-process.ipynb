## 3. 处理场馆数据

```python
stadium = pd.read_excel('./database/xlsx/场馆信息.xlsx')
stadium.head(1)
```

```python
stadium.isnull().sum()  # 检查缺失值
```

```python
from pandas import DataFrame
stadium_processed = DataFrame()
stadium_processed['场馆名称'] = stadium['标题']
stadium_processed['文档网址'] = stadium['标题链接']
stadium_processed['浏览量'] = stadium['标签'].apply(lambda x: int(str(x).replace('',''))) # 将浏览量转换为整数
stadium_processed['场馆地址'] = stadium['text'].apply(lambda x: str(x).replace('',''))
stadium_processed['场馆介绍'] = stadium['字段1'].apply(
    lambda x: str(x if x is not None and x is not np.nan else '暂无').replace('',''))
stadium_processed['开馆时间'] = stadium['字段2'].apply(lambda x: str(x).replace('开馆时间：',''))
stadium_processed['联系电话'] = stadium['字段3'].apply(lambda x: str(x).replace('联系电话：',''))
stadium_processed['场馆面积'] = stadium['字段5'].apply(lambda x: str(x if x is not None and x is not np.nan else '暂无'))
stadium_processed['容纳人数'] = stadium['字段6'].apply(lambda x: str(x if x is not None and x is not np.nan else '暂无'))
stadium_processed['温馨提示'] = stadium['字段7'].apply(lambda x: str(x if x is not None and x is not np.nan else '暂无'))
```

```python
stadium_processed.info  # 显示前5行数据
```

```python
stadium_processed.to_json('./database/json/stadium_data.jsonl', orient='records', force_ascii=False, lines=False)  # 保存为JSONL文件
```
